                                   IMPORTING THE LIBRARIES

```python
import os
from roboflow import Roboflow
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

                                DOWNLOAD DATASET FROM THE ROBFLOW

```python
rf = Roboflow(api_key="YOUR_API_KEY_HERE")
project = rf.workspace("shyams-workspace-i8nfl").project("motorcycle-helmet-7d2gr-mugjt")
version = project.version(1)
dataset = version.download("yolov8")
print("Dataset location:", dataset.location)
```

                                         TRAIN THE MODEL

```python
model = YOLO("yolov8n.pt")  
model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="helmet_violation_detector",
    patience=10
)
print("Training complete!") 
```

                                        EVALUATE THE MODEL

```python
metrics = model.val()
print(f"mAP50: {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")
print(f"Precision: {metrics.box.p:.3f}")
print(f"Recall: {metrics.box.r:.3f}")
```

                                       SAVE TRAINING RESULTS

```python
import shutil
results_src = "runs/detect/helmet_violation_detector"
print("Training results saved in:", results_src)
print("Files:", os.listdir(results_src))
```
